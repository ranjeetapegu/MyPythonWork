```python
# Importing Libraries and csv file

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
Movies =pd.read_csv("Movie-Ratings.csv")
Movies.columns=["Film","Genre","CriticRating","AudienceRating","BudgetMillions","Year"]
# Categorical Variables**
Movies.Year = Movies.Year.astype('category')
Movies.Genre = Movies.Genre.astype('category')
Movies.Film = Movies.Film.astype('category')
Movies.head()
```

```python
# define a empty list
year=[]
Mylegend =[]
for i in Movies.Year.unique():
    Mylegend.append(i)
    year.append(Movies[Movies.Year ==  i].BudgetMillions)
 
# print(year)
plt.hist(year,histtype='bar',rwidth=1,orientation='vertical', \
        stacked=False, label =Mylegend)
plt.legend()
plt.show()
```

```python
# define a empty list

Genre=[]
Mylegend =[]
for i in Movies.Genre.unique():
    Mylegend.append(i)
    Genre.append(Movies[Movies.Genre ==  i].BudgetMillions)

sns.set_style('white')
fig,ax =plt.subplots()
fig.set_size_inches(10,8)
plt.hist(Genre,bins=30,histtype='bar',rwidth=1, \
         stacked=True, ec='white',label =Mylegend)
plt.title('Movie Budget Distributions in Millions',color='darkblue',fontname='Arial', fontsize=35)
plt.xlabel('Number of Movies', fontsize=15)
plt.ylabel('Budget in Millions', fontsize=15)
plt.legend(frameon=True, fancybox=True, prop={'size':12})
plt.show()
```
