```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
Movies = pd.read_csv('Movies_2.csv',encoding = "ISO-8859-1")
```

```python

Movies.columns=['WeekDay','Director','Genre','MovieTitle','ReleaseDate','Studio',
                'AdjustedGrossMillions','BudgetMillions','GrossMillions','IMDbRating',\
                'MovieLensRating','OverseasMillions','OverseasPercent','ProfitMillions',\
                'ProfitPercent','RuntimeMins','USMillions', 'GrossPercentUS']

Movies.WeekDay= Movies.WeekDay.astype('category')
Movies.Genre = Movies.Genre.astype('category')
Movies.Studio = Movies.Studio.astype('category')
```

```python
#filtering the Data
sns.set_style("whitegrid")

Mov=pd.DataFrame(Movies[Movies['Genre'].isin(['action',\
                'comedy','adventure', 'animation','drama'])])
Mov1=pd.DataFrame(Mov[Mov['Studio'].isin(['Buena Vista Studios',\
                   'Sony','Universal','WB','Paramount Pictures','Fox'])]) 
Mov1.to_csv('out.csv')
mv = pd.read_csv('out.csv')
mv.Genre.unique()
```

```python
sns.set_style("darkgrid")
fig,ax=plt.subplots()
fig.set_size_inches(11,10)

fig1 = sns.boxplot(data=mv, x='Genre', y='GrossPercentUS',
                 orient='v', color='lightgray', showfliers=False)
plt.setp(fig1.artists, alpha=0.5)

# Add in points to show each observation
sns.stripplot(x='Genre', y='GrossPercentUS',
              data=mv, jitter=True, size=6, 
              linewidth=0, 
              hue = 'Studio', alpha=0.7)

fig1.axes.set_title('Domestic Gross % Genre',fontsize=30)
fig1.set_xlabel('Genre',fontsize=20)
fig1.set_ylabel('Gross % US',fontsize=20)

# Define where to place the legend
fig1.legend(bbox_to_anchor=(1.05, 1), loc=0, borderaxespad=0.)
plt.show()
```

```python
ax = sns.boxplot(data=mv, x='Genre', y='GrossPercentUS',
           orient='v', color='pink', showfliers=False)

#plt.setp(ax.artists, alpha=0.5)
sns.swarmplot(data=mv, x='Genre', y='GrossPercentUS',hue='Studio');
ax.axes.set_title('Domestic Gross % Genre',fontsize=30)
ax.set_xlabel('Genre',fontsize=20)
ax.set_ylabel('Gross % US',fontsize=20)
ax.legend(bbox_to_anchor=(1.05, 1),loc=0, borderaxespad=0.)
plt.show()
```
