```python
# Importing Libraries and csv file

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
Movies =pd.read_csv("Movie-Ratings.csv")
Movies.columns=["Film","Genre","CriticRating","AudienceRating","BudgetMillions","Year"]
# Categorical Variables**
Movies.Year = Movies.Year.astype('category')
Movies.Genre = Movies.Genre.astype('category')
Movies.Film = Movies.Film.astype('category')
Movies.info()
```

```python
# JoinPLot 

J= sns.jointplot(Movies.AudienceRating,Movies.CriticRating,color='g',
                 kind='scatter', size=6,s= 50)
```

```python
J = sns.jointplot(Movies.AudienceRating, Movies.BudgetMillions, color='c', kind='kde')
```

```python
J =sns.jointplot(Movies.CriticRating, Movies.AudienceRating, color='r',kind='hex')
```

```python
J =sns.jointplot(Movies.CriticRating, Movies.AudienceRating, color='y',kind='reg')
```
